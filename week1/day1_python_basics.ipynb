```python
#Functions

def calculate_average(numbers):
  total=sum(numbers)
  count=len(numbers)
  return total / count

marks= [55,70,82,90,65]

avg=calculate_average(marks)
print("Average",avg)
```

```python
marks = [55,70,82,90,65,40,95]

passed_students = []

for mark in marks:
  if mark >= 60:
    passed_students.append(mark)
print("passed_students:",passed_students)
print("number of students passed:",len(passed_students))

def average(numbers):
  total = sum(numbers)
  count = len(numbers)
  return total/count

avg=average(passed_students)
print("Passed students average:",avg)
```
