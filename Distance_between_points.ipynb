```python
A=(5,3)
B=(2,4)
```

```python
from scipy.spatial import distance
```

Euclidean Distance

```python
d = distance.euclidean(A,B)
print(d)
```

Manhattan Distance

```python
d= distance.cityblock(A,B)
print(d)
```

Cosine Distance

```python
d = distance.cosine(A,B)
print(d)
```
